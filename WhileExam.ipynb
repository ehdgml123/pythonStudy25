```python
# 1회용이 아닌 반복문을 활용해서 성적 처리 프로그램을 만들어 보자.
# 성적에 CRUD 를 구현한다.
# 1번 메뉴를 선택하면 성적을 입력한다.
# 2번 메뉴를 선택하면 성적을 확인한다.
# 4번 메뉴를 선택하면 성적을 삭제한다.
# 5번 메뉴를 선택하면 프로그램을 종료한다.

run = True  # while문의 반복을 처리하기 위함 False 로 바뀌면 반복을 종료
while run :
    prompt = """
    1. 성적추가
    2. 성적확인
    3. 성적수정
    4. 성적삭제
    5. 프로그램 종료

    Enter number(1 ~ 5) : """
    print(prompt)

    select = int(input("번호 입력 :"))
    if select == 1 :
        print("성적추가")
    elif select == 2 :
        print("성적을 확인합니다")
    elif select == 3 :
        print("성적을 수정합니다.")
    elif select == 4 :
        print("성적을 삭제합니다.")
    elif select == 5 :
        print("프로그램을 종료합니다.")
        run = False
```

```python
run = True

while run :
     prompt = """
        1. 성적추가하기
        2 .성적확인하기
        3. 성적 수정하기
        4. 성적 삭제하기
        5. 프로그램 종료하기
        Enter number(1 ~ 5) :
     """
     print(prompt)

     select = int(input("번호 입력 :"))
     if select == 1 :
        print("성적 추가하기")
        kor = int(input("수학 :" ))
        eng = int(input("영어 :" ))
        si = int(input("과학 :" ))
     elif select == 2 :
        print("성적을 확인합니다")
        print("수학 : %d , 영어 : %d , 과학 : %d" % (kor, eng, si))
     elif select == 3 :
        print("성적을 수정합니다.")
        prompt1 = """
           1. 수학
           2. 영어
           3. 과학
           Enter number(1 ~ 3) :
        """
        select1 = int(input(prompt1))
        if select1 == 1 :
            kor = int(input("수학 :" ))
        elif select1 == 2 :
            eng = int(input("영어 :" ))
        elif select1 == 3 :
            si = int(input("과학 :" ))

        elif select == 4 :
         print("성적 삭제하기")
         kor = 0
         eng = 0
         si = 0
     elif select == 5 :
        print("프로그램을 종료합니다.")
        run = False
```
